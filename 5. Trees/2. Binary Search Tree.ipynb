```python
class BinaryTreeNode:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
        
def takeInput():
    inputList = [int(i) for i in input().split()]
    q = []
    
    root = BinaryTreeNode(inputList[0])
    q.append(root)
    
    i = 1
    while q:
        node = q.pop(0)
        
        if inputList[i] != -1:
            left = BinaryTreeNode(inputList[i])
            node.left = left
            q.append(left)
            
        i += 1
        
        if inputList[i] != -1:
            right = BinaryTreeNode(inputList[i])
            node.right = right
            q.append(right)
        
        i += 1
        
    return root

def printTree(root, detailed=True):
    q = []
    
    q.append(root)
    if detailed:
        while q:
            node = q.pop(0)
            print(node.data, end = ': ')

            if node.left:
                print('L:'+str(node.left.data), end = ', ')
                q.append(node.left)

            if node.right:
                print('R:'+str(node.right.data), end = '')
                q.append(node.right)

            print()
            
    else:
        while q:
            node = q.pop(0)
            print(node.data, end = ' ')
            
            if node.left:
                q.append(node.left)
                
            if node.right:
                q.append(node.right)
```

# BST - 1

```python
def searchInBST(root, x):
    if not root:
        return False
    
    if x < root.data:
        return searchInBST(root.left, x)
    
    if x > root.data:
        return searchInBST(root.right, x)
    
    return True
```

```python
# print nodes in range K1 and K2

def printNodes(root, k1, k2):
    if not root:
        return
    
    if k1 <= root.data:
        printNodes(root.left, k1, k2)
        
    if k1 <= root.data and root.data <= k2:
        print(root.data, end = ' ')
        
    if root.data <= k2:
        printNodes(root.right, k1, k2)
```

```python
# construct a BST from sorted array

def buildBST(arr):
    if not arr:
        return None
    
    mid = len(arr)//2
    root = BinaryTreeNode(arr[mid])
    
    left = buildBST(arr[: mid])
    right = buildBST(arr[mid+1: ])
    
    root.left = left
    root.right = right
    
    return root
```

```python
# check the given tree is BST

def isBST(root):
    if not root:
        return 100000, -100000, True
    
    lmin, lmax, lbst = isBST(root.left)
    rmin, rmax, rbst = isBST(root.right)
    
    if lmax < root.data and root.data < rmin and lbst and rbst:
        return min(lmin, root.data), max(rmax, root.data), True
    
    return min(lmin, root.data), max(rmax, root.data), False

def isBST1(root, min_=-100000, max_=100000):
    if not root:
        return True
    
    if min_ > root.data or root.data > max_:
        return False
    
    return isBST1(root.left, min_, root.data) and isBST1(root.right, root.data, max_)
```

# BST - 2

```python
# find path root to given node

def path(root, node):
    if not root:
        return
    
    if root.data == node:
        return [root.data]
    
    left = path(root.left, node)
    if left:
        left.append(root.data)
        return left
    
    right = path(root.right, node)
    if right:
        right.append(root.data)
        return right
```

```python
# find path in BST

def bstPath(root, node):
    if not root:
        return []
    
    if node < root.data:
        left = bstPath(root.left, node)
        if left:
            left.append(root.data)
        return left
    
    if node > root.data:
        right = bstPath(root.right, node)
        if right:
            right.append(root.data)
        return right
    
    return [root.data]
```

# BST Class

```python
class BinaryTreeNode:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
        
class BST:
    def __init__(self):
        self.root = None
        self.count = 0
        
    def insertHelper(self, root, data):
        if not root:
            return BinaryTreeNode(data)
        
        if data < root.data:
            root.left = self.insertHelper(root.left, data)
            
        if data > root.data:
            root.right = self.insertHelper(root.right, data)
            
        return root
    
    def insert(self, data):
        self.root = self.insertHelper(self.root, data)
        self.count += 1
    
    def getMin(self, root):
        if not root:
            return 100000
        
        left = self.getMin(root.left)
        right = self.getMin(root.right)
        
        return min(left, right, root.data)
    
    def deleteHelper(self, root, data):
        if not root:
            return root, False
        
        if data < root.data:
            root.left, deleted = self.deleteHelper(root.left, data)
            return root, deleted
            
        if data > root.data:
            root.right, deleted = self.deleteHelper(root.right, data)
            return root, deleted
            
        if not root.left and not root.right:
            return None, True
        
        if not root.left:
            return root.right, True
        
        if not root.right:
            return root.left, True
        
        rightMin = self.getMin(root.right)
        root.data = rightMin
        root.right, deleted = self.deleteHelper(root.right, rightMin)
        
        return root, deleted
    
    def delete(self, data):
        self.root, deleted = self.deleteHelper(self.root, data)
        if deleted:
            self.count -= 1
    
    def searchHelper(self, root, data):
        if not root:
            return False
        
        if data < root.data:
            return self.searchHelper(root.left, data)
        
        if data > root.data:
            return self.searchHelper(root.right, data)
        
        return True
    
    def search(self, data):
        return self.searchHelper(self.root, data)
    
    def printTreeHelper(self, root, detailed):
        q = []
    
        q.append(root)
        if detailed:
            while q:
                node = q.pop(0)
                print(node.data, end = ': ')

                if node.left:
                    print('L:'+str(node.left.data), end = ', ')
                    q.append(node.left)

                if node.right:
                    print('R:'+str(node.right.data), end = '')
                    q.append(node.right)

                print()

        else:
            while q:
                node = q.pop(0)
                print(node.data, end = ' ')

                if node.left:
                    q.append(node.left)

                if node.right:
                    q.append(node.right)
    
    def printTree(self, detailed=True):
        self.printTreeHelper(self.root, detailed)
```
