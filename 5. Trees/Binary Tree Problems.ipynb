```python
class BinaryTreeNode:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
        
def takeInput():
    inputList = [int(i) for i in input().split()]
    q = []
    
    root = BinaryTreeNode(inputList[0])
    q.append(root)
    
    i = 1
    while q:
        node = q.pop(0)
        
        if inputList[i] != -1:
            left = BinaryTreeNode(inputList[i])
            node.left = left
            q.append(left)
            
        i += 1
        
        if inputList[i] != -1:
            right = BinaryTreeNode(inputList[i])
            node.right = right
            q.append(right)
        
        i += 1
        
    return root

def printTree(root, detailed=True):
    q = []
    
    q.append(root)
    if detailed:
        while q:
            node = q.pop(0)
            print(node.data, end = ': ')

            if node.left:
                print('L:'+str(node.left.data), end = ', ')
                q.append(node.left)

            if node.right:
                print('R:'+str(node.right.data), end = '')
                q.append(node.right)

            print()
            
    else:
        while q:
            node = q.pop(0)
            print(node.data, end = ' ')
            
            if node.left:
                q.append(node.left)
                
            if node.right:
                q.append(node.right)
```

```python
# no of nodes

def countNodes(root):
    if not root:
        return 0
    
    return 1 + countNodes(root.left) + countNodes(root.right)
```

```python
# sum of nodes

def nodesSum(root):
    if not root:
        return 0
    
    return root.data + nodesSum(root.left) + nodesSum(root.right)
```

```python
# node with largest data

def nodeWithLargestData(root):
    if not root.left and not root.right:
        return root
    
    largest_node = root
    if root.left:
        left_largest = nodeWithLargestData(root.left)
        if left_largest.data > largest_node.data:
            largest_node = left_largest
        
    if root.right:
        right_largest = nodeWithLargestData(root.right)
        if right_largest.data > largest_node.data:
            largest_node = right_largest
        
    return largest_node
```

```python
# no of nodes greater than x

def countNodesGreaterThanX(root, x):
    if not root:
        return 0
    
    if root.data > x:
        return 1 + countNodesGreaterThanX(root.left, x) + countNodesGreaterThanX(root.right, x)
    
    return countNodesGreaterThanX(root.left, x) + countNodesGreaterThanX(root.right, x)
```

```python
# height of tree

def height(root):
    if not root:
        return 0
    
    return 1 + max(height(root.left), height(root.right))
```

```python
# no of leaf nodes

def countLeafNodes(root):
    if not root:
        return 0
    
    if not root.left and not root.right:
        return 1
    
    return countLeafNodes(root.left) + countLeafNodes(root.right)
```

```python
# print nodes at level k

def printNodesAtLevelK(root, k):
    if root is None or k < 1:
        return
    
    if k == 1:
        print(root.data, end = ' ')
        return
    
    printNodesAtLevelK(root.left, k - 1)
    printNodesAtLevelK(root.right, k - 1)
```

```python
# print nodes at depth K

def printNodesAtDepthK(root, k):
    if not root or k < 0:
        return
    
    if k == 0:
        print(root.data, end=' ')
        return
    
    printNodesAtDepthK(root.left, k-1)
    printNodesAtDepthK(root.right, k-1)
```

```python
# replace nodes with depth

def replaceNodesWithDepth(root, currDepth=0):
    if not root:
        return
    
    root.data = currDepth
    replaceNodesWithDepth(root.left, currDepth+1)
    replaceNodesWithDepth(root.right, currDepth+1)
    
    return root
```

```python
# check whether a node exists or not

def isPresent(root, node):
    if not root:
        return False
    
    if root.data == node:
        return True
    
    return isPresent(root.left, node) or isPresent(root.right, node)
```

```python
# print nodes without siblings

def nodesWithoutSiblings(root):
    if not root or not root.left and not root.right:
        return
    
    if not root.right:
        print(root.left.data, end=' ')
        
    if not root.left:
        print(root.right.data, end=' ')
        
    nodesWithoutSiblings(root.left)
    nodesWithoutSiblings(root.right)
```

```python
# remove leaf nodes

def removeLeafNodes(root):
    if not root or (not root.left and not root.right):
        return None
    
    root.left = removeLeafNodes(root.left)
    root.right = removeLeafNodes(root.right)
    
    return root
```

```python
# mirror binary tree - update binary tree with its corresponding mirror image

def mirrorBinaryTree(root):
    if not root:
        return root
    
    left_mirror = mirrorBinaryTree(root.left)
    root.left = mirrorBinaryTree(root.right)
    root.right = left_mirror
    
    return root
```

```python
# check if binary tree is balanced

def height(root):
    if not root:
        return 0
    
    return 1 + max(height(root.left), height(root.right))

def isBalanced(root):
    if not root:
        return True
    
    if abs(height(root.left) - height(root.right)) > 1:
        return False
    
    return isBalanced(root.left) and isBalanced(root.right)

def isBalanced(root):
    if not root:
        return 0, True
    
    lh, left_res = isBalanced(root.left)
    rh, right_res = isBalanced(root.right)
    
    if abs(lh - rh) > 1:
        return 1 + max(lh, rh), False
    
    return 1 + max(lh, rh), left_res and right_res
```

```python
# diameter of a binary tree

def height(root):
    if not root:
        return 0
    
    return 1 + max(height(root.left), height(root.right))

def diameter(root):
    if not root:
        return 0
    
    ld = diameter(root.left)
    rd = diameter(root.right)
    
    d = max(ld, rd)
    lh = height(root.left)
    rh = height(root.right)
    if d > 1 + lh + rh:
        return d
    
    return 1 + lh + rh

def diameter(root):
    if not root:
        return 0, 0
    
    ld, lh = diameter(root.left)
    rd, rh = diameter(root.right)
    
    d = max(ld, rd)
    if d > 1 + lh + rh:
        return d, 1 + max(lh, rh)
    
    return 1 + lh + rh, 1 + max(lh, rh)
```

```python
# create & insert duplicates

def insertDuplicates(root):
    if not root:
        return root
    
    duplicate = BinaryTreeNode(root.data)
    duplicate.left = insertDuplicates(root.left)
    root.left = duplicate
    root.right = insertDuplicates(root.right)
    
    return root
```

```python
# get min and max

def minMax(root):
    if not root:
        return 100000, -100000
    
    lmin, lmax= minMax(root.left)
    rmin, rmax = minMax(root.right)
    
    return min(root.data, lmin, rmin), max(root.data, lmax, rmax)
```

```python
# root-to-leaf paths where the sum of all the node data along the path is equal to K.

def returnRootToLeafPathsSumToK(root, k):
    if not root:
        return []
    
    if not root.left and not root.right:
        if k == root.data:
            return [[root.data]]
        return []
    
    lpaths = returnRootToLeafPathsSumToK(root.left, k-root.data)
    rpaths = returnRootToLeafPathsSumToK(root.right, k-root.data)
    
    paths = []
    for path in lpaths:
        path.insert(0, root.data)
        paths.append(path)

    for path in rpaths:
        path.insert(0, root.data)
        paths.append(path)

    return paths


def printRootToLeafPathsSumToK(root, k, path=''):
    if not root:
        return
    
    if not root.left and not root.right:
        if k == root.data:
            print(path+str(root.data))
        return
    
    printRootToLeafPathsSumToK(root.left, k-root.data, path+str(root.data)+' ')
    printRootToLeafPathsSumToK(root.right, k-root.data, path+str(root.data)+' ')
```

```python
# Print nodes at distance K from node

def printNodesAtDepthK(root, k):
    if not root:
        return
    
    if k == 0:
        print(root.data, end = ' ')
        return
    
    printNodesAtDepthK(root.left, k-1)
    printNodesAtDepthK(root.right, k-1)
    
def printNodesAtDisK(root, node, k):
    if not root:
        return -1
    
    if root.data == node:
        printNodesAtDepthK(root, k)
        return 0
    
    ld = printNodesAtDisK(root.left, node, k)
    if ld != -1:
        if ld+1 == k:
            print(root.data, end = ' ')
        else:
            printNodesAtDepthK(root.right, k-(ld+1)-1)
        return ld+1
        
    rd = printNodesAtDisK(root.right, node, k)
    if rd != -1:
        if rd+1 == k:
            print(root.data, end = ' ')
        else:
            printNodesAtDepthK(root.left, k-(rd+1)-1)
        return rd+1
    
    return -1
```

```python
# check cousins
'''Two nodes are said to be cousins of each other if they are at same level of the Binary Tree and have different parents.'''

def parent(root, x):
    if not root or not root.left and not root.right:
        return -1
        
    if root.left and root.left.data == x or root.right and root.right.data == x:
        return root
        
    left_parent = parent(root.left, x)
    return left_parent if left_parent != -1 else parent(root.right, x)

def level(root, x):
    if not root:
        return -1
    
    if root.data == x:
        return 1
    
    level_left = level(root.left, x)
    if level_left != -1:
        return 1+level_left
    
    level_right = level(root.right, x)
    if level_right != -1:
        return 1+level_right
    
    return -1

def areCousins(root, a, b):
    return True if parent(root, a) != parent(root, b) and level(root, a) == level(root, b) else False
```

```python
# return the longest path from leaf to root

def longestPathLeafToRoot(root):
    if not root:
        return []
    
    left_longest_path = longestPathLeafToRoot(root.left)
    right_longest_path = longestPathLeafToRoot(root.right)
    
    if len(left_longest_path) > len(right_longest_path):
        left_longest_path.append(root.data)
        return left_longest_path
    right_longest_path.append(root.data)
    return right_longest_path
```

```python
# 21: Check if two binary trees are identical

def isIdentical(root1, root2):
    if not root1 and not root2:
        return True
    
    if not root1 or not root2:
        return False
    
    return root1.data == root2.data and isIdentical(root1.left, root2.left) and isIdentical(root1.right, root2.right)
```

```python
# 23: print the top view

def topView(root):
    h = {}
    q = [(root, 0)]
    
    while q:
        node, line = q.pop(0)
        if line not in h:
            h[line] = node
            
        if node.left:
            q.append((node.left, line-1))
            
        if node.right:
            q.append((node.right, line+1))
            
    l = sorted(h)
    for i in l:
        node = h[i]
        print(node.data, end = ' ')
```

```python
# 26: Print Vertical Order of Binary Tree

def printVerticalOrder(root):
    h = {}
    q = [(root, 0)]
    while q:
        node, line = q.pop(0)
        if line in h:
            h[line].append(node.data)
            
        else:
            h[line] = [node.data]
            
        if node.left:
            q.append((node.left, line-1))
            
        if node.right:
            q.append((node.right, line+1))
            
    l = sorted(h)
    for i in l:
        print(*h[i])
```

```python
# 30: Print Binary Tree in Spiral Order

def spiralOrder(root):
    s1, s2 = [root], []
    
    while s1 or s2:
        while s1:
            root = s1.pop()
            print(root.data, end = ' ')
            
            if root.right:
                s2.append(root.right)
                
            if root.left:
                s2.append(root.left)
                
        print()
        while s2:
            root = s2.pop()
            print(root.data, end = ' ')
            
            if root.left:
                s1.append(root.left)
                
            if root.right:
                s1.append(root.right)
        
        print()
```

```python
# 32. get level with max width and width also

def maxLevelWithWidth(root):
    level = 1
    tempLevel = 1
    width = 1
    q = [root]
    while q:
        count = len(q)
        if count > width:
            width = count
            level = tempLevel
            
        while count:
            root = q.pop(0)
            
            if root.left:
                q.append(root.left)
                
            if root.right:
                q.append(root.right)
            
            count -= 1
            
        tempLevel += 1
        
    return level, width
```

```python
# 33. check if trees are mirror trees to each other

def ifTreesSame(root1, root2):
    if root1 is None and root2 is None:
        return True
    
    if root1 is None or root2 is None:
        return False
    
    return root1.data == root2.data and ifTreesSame(root1.left, root2.left) and ifTreesSame(root1.right, root2.right)

def ifMirrorTrees(root1, root2):
    if root1 is None and root2 is None:
        return True
    
    if root1 is None or root2 is None:
        return False
    
    return root1.data == root2.data and ifMirrorTrees(root1.left, root2.right) and ifMirrorTrees(root1.right, root2.left)
```

```python
# 36: Check if Binary Tree is Foldable Tree

def mirrorStructure(root1, root2):
    if not root1 and not root2:
        return True
    
    if not root1 or not root2:
        return False
    
    return mirrorStructure(root1.left, root2.right) and mirrorStructure(root1.right, root2.left)

def isFodable(root):
    return mirrorStructure(root.left, root.right)
```
