height checker -
https://leetcode.com/problems/height-checker

```python
def nextGreaterEelements1(arr):
    s = []
    h = {}
    for i in range(len(arr)-1, -1, -1):
        while s and s[-1] <= arr[i]:
            s.pop()
            
        if not s:
            h[arr[i]] = -1
            
        else:
            h[arr[i]] = s[-1]
            
        s.append(arr[i])
        
    return h

arr = [2, 10, 12, 1, 11]
nextGreaterEelements1(arr)
```

```python
def nextGreaterEelements2(arr):
    s = []
    h = {}
    for i in range((2*len(arr))-1, -1, -1):
        while s and s[-1] <= arr[i%len(arr)]:
            s.pop()
            
        if i < len(arr):
            if not s:
                h[arr[i]] = -1
                
            else:
                h[arr[i]] = s[-1]
            
        s.append(arr[i%len(arr)])
        
    return h

arr = [2, 10, 12, 1, 11]
nextGreaterEelements2(arr)
```
