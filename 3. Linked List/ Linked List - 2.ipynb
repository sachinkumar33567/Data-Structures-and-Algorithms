```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
def takeInput():
    lis = [int(i) for i in input().split()]
    
    head = None
    tail = None
    
    for i in lis:
        if i == -1:
            break
            
        if not head:
            head = Node(i)
            tail = head
            
        else:
            tail.next = Node(i)
            tail = tail.next
            
    return head

def printLL(head):
    if not head:
        return
    
    print(head.data, end = ' ')
    printLL(head.next)
```

## merge sort on LL

```python
def mergeTwoSortedLL(h1, h2):
    if not h1:
        return h2
    
    if not h2:
        return h1
    
    head = None
    
    if h1.data < h2.data:
        head = h1
        head.next = mergeTwoSortedLL(h1.next, h2)
    else:
        head = h2
        head.next = mergeTwoSortedLL(h1, h2.next)
        
    return head

def middle(head):
    prev = None
    slow = head
    fast = head
    
    while fast and fast.next:
        prev = slow
        slow = slow.next
        fast = fast.next.next
        
    return prev

def mergeSort(head):
    if not head or not head.next:
        return head
    
    mid = middle(head)
    next_mid = mid.next
    mid.next = None
    
    left = mergeSort(head)
    right = mergeSort(next_mid)
    
    return mergeTwoSortedLL(left, right)
```

```python
# Even after Odd LinkedList

def evenAfterOdd(head):
    eh = et = None
    oh = ot = None
    
    while head:
        if head.data % 2:
            if not oh:
                oh = head
                ot = oh
            else:
                ot.next = head
                ot = ot.next
        else:
            if not eh:
                eh = head
                et = head
            else:
                et.next = head
                et = et.next
        head = head.next
        
    if not oh:
        return eh
    if not eh:
        return oh
    
    et.next = None
    ot.next = eh
    
    return oh
```

```python
# Delete every N nodes
'''Traverse the linked list such that you retain the 'M' nodes, then delete the next 'N' nodes.'''

def skipMdeleteN(head, m, n):
    if m < 1:
        return None
    
    curr = head
    while curr:
        skip = m
        while skip-1 and curr:
            curr = curr.next
            skip -= 1
            
        if not curr:
            return head
        
        delete = n
        ptr = curr.next
        while delete and ptr:
            ptr = ptr.next
            delete -= 1
            
        curr.next = ptr
        curr = ptr
        
    return head
```

```python
# Swap two Nodes of LL

def swap(head, ith, jth):
    if ith > jth:
        ith, jth = jth, ith
        
    prev1 = None
    curr1 = head
    i, j = ith, jth
    while i:
        prev1 = curr1
        curr1 = curr1.next
        i -= 1
        j -= 1
        
    if j == i+1:
        curr2 = curr1.next
        temp = curr2.next
        if ith == 0:
            curr2.next = curr1
            curr1.next = temp
            head = curr2
            return head
        
        prev1.next = curr2
        curr2.next = curr1
        curr1.next = temp
        return head
    
    prev2 = prev1
    curr2 = curr1
    while j:
        prev2 = curr2
        curr2 = curr2.next
        j -= 1
        
    temp = curr2.next
    if ith == 0:
        curr2.next = curr1.next
        prev2.next = curr1
        curr1.next = temp
        head = curr2
        return head
    
    prev1.next = curr2
    curr2.next = curr1.next
    prev2.next = curr1
    curr1.next = temp
    
    return head
```

```python
# kReverse
'''reverse the nodes of the linked list 'k' at a time and return its modified list.'''

def reverse(head):
    if not head or not head.next:
        return head
    
    recurHead = reverse(head.next)
    tail = head.next
    head.next = None
    tail.next = head
    
    return recurHead

def kReverse(head, k):
    if not head:
        return head
    
    curr = head
    cnt = k-1
    while cnt > 0 and curr.next:
        curr = curr.next
        cnt -= 1
        
    nextLL = curr.next
    curr.next = None
    recurHead = kReverse(nextLL, k)
    
    newHead = reverse(head)
    head.next = recurHead
    return newHead
```

## bubble sort

```python
def length(head):
    if not head:
        return 0
    
    return 1 + length(head.next)

def bubbleSort(head):
    n = length(head)
    while n-1:
        prev = None
        curr = head
        next = head.next
        while next:
            if curr.data > next.data:
                if not prev:
                    temp = next.next
                    next.next = curr
                    curr.next = temp
                    head = next
                else:
                    temp = next.next
                    prev.next = next
                    next.next = curr
                    curr.next = temp
                
                prev = next
                next = curr.next
            
            else:
                prev = curr
                curr = next
                next = next.next
                
        n -= 1
        
    return head
```
