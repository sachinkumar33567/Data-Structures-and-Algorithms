```python
import random

def print_mat(mat):
    for i in mat:
        print(i)
        
def get_curr_state(mat):
    # if any where 2048 is present
    for row in range(4):
        for col in range(4):
            if mat[row][col] == 64:
                return 'you won'
            
    # if any where 0 if present
    for row in range(4):
        for col in range(4):
            if not mat[row][col]:
                return 'game is not over'
            
    # if any two same values are present
    # checking horizontally and vertically
    for row in range(3):
        for col in range(3):
            if mat[row][col] == mat[row][col+1] or mat[row][col] == mat[row+1][col]:
                return 'game is not over'
            
    # checking in last column
    for row in range(3):
        if mat[row][3] == mat[row+1][3]:
            return 'game is not over'
        
    # checking in last row
    for col in range(3):
        if mat[3][col] == mat[3][col+1]:
            return 'game is not over'
        
    return 'you lost'

def transpose(mat):
    '''make rows to the columns'''
    new = [[0 for _ in range(4)] for _ in range(4)]
    
    for row in range(4):
        for col in range(4):
            new[col][row] = mat[row][col]
            
    return new

def reverse(mat):
    new = [[0 for _ in range(4)] for _ in range(4)]
    
    for row in range(4):
        for col in range(3, -1, -1):
            new[row][3-col] = mat[row][col]
            
    return new

def merge(mat):
    '''left merge'''
    merge_movement = False
    for row in range(4):
        for col in range(3):
            if mat[row][col] and mat[row][col] == mat[row][col+1]:
                mat[row][col] += mat[row][col+1]
                mat[row][col+1] = 0
                merge_movement = True
                
    return mat, merge_movement

def compress(mat):
    '''left compression'''
    new = [[0 for _ in range(4)] for _ in range(4)]
    compress_movement = False
    
    for row in range(4):
        pos = 0
        for col in range(4):
            if mat[row][col]:
                new[row][pos] = mat[row][col]
                if pos != col:
                    compress_movement = True
                pos += 1
                
    return new, compress_movement

def add_new_2(mat):
    row = random.randint(0, 3)
    col = random.randint(0, 3)
    
    while mat[row][col]:
        row = random.randint(0, 3)
        col = random.randint(0, 3)
        
    mat[row][col] = 2

def move_down(mat):
    transposed = transpose(mat)
    right_moved, movement = move_right(transposed)
    transposed = transpose(right_moved)
    
    return transposed, movement

def move_up(mat):
    transposed = transpose(mat)
    left_moved, movement = move_left(transposed)
    transposed = transpose(left_moved)
    
    return transposed, movement

def move_right(mat):
    reversed_ = reverse(mat)
    left_moved, movement = move_left(reversed_)
    reversed_ = reverse(left_moved)
    
    return reversed_, movement

def move_left(mat):
    compressed, compress_movement = compress(mat)
    merged, merge_movement = merge(compressed)
    compressed = compress(merged)[0]
    
    movement = compress_movement or merge_movement
    
    return compressed, movement

def start_game():
    mat = [[0 for _ in range(4)] for _ in range(4)]
    add_new_2(mat)
    add_new_2(mat)
    return mat
```

```python
mat = start_game()
print_mat(mat)

status = get_curr_state(mat)
while status == 'game is not over':
    command = input()
    
    if command == 'j':
        mat, movement = move_left(mat)
        
    elif command == 'l':
        mat, movement = move_right(mat)
        
    elif command == 'i':
        mat, movement = move_up(mat)
        
    elif command == 'k':
        mat, movement = move_down(mat)
        
    else:
        print('game terminated')
        break
        
    if movement:
            add_new_2(mat)
    print_mat(mat)
    status = get_curr_state(mat)
    
print(status)
```
