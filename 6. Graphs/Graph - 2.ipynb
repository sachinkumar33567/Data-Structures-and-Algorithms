# Kruskal's Algorithm

```python
# union find algorithm

def find(i):
    if i == parent[i]:
        return i
    return find(parent[i])

def union(v1, v2):
    parent[find(v2)] = find(v1)
```

```python
# union by rank and path compression

def find(i):
    if i == parent[i]:
        return i
    parent[i] = find(parent[i])
    return parent[i]

def union(v1, v2):
    if rank[find(v1)] > rank[find(v2)]:
        parent[find(v2)] = find(v1)
        
    elif rank[find(v1)] < rank[find(v2)]:
        parent[find(v1)] = find(v2)
        
    else:
        parent[find(v2)] = find(v1)
        rank[find(v1)] += 1
```

```python
def kruskal(v, edges):
    edges.sort()
    global parent, rank
    
    parent = [i for i in range(v)]
    rank = [0 for _ in range(v)]
    mst = []
    
    i = 0
    while v - 1:
        edge = edges[i]
        v1, v2, wt = edge[1], edge[2], edge[0]
        if find(v1) != find(v2):
            union(v1, v2)
            mst.append([v1, v2, wt])
            
        i += 1
        v -= 1
            
    return mst
```

```python
v, e = input().split()

edges = []
for _ in range(int(e)):
    v1, v2, wt = input().split()
    edges.append([int(wt), int(v1), int(v2)])
    
mst = kruskal(int(v), edges)
```

# Prim's Algorithm

```python
def getMinWeightedVertex(wt, unvisited):
    v = unvisited[0]
    minimum = wt[v]
    for i in unvisited:
        if wt[i] < minimum:
            minimum = wt[i]
            v = i
            
    return v

def prims(adjMatrix, nVertices):
    parent = [-1 for _ in range(nVertices)]
    wt = [100000 for _ in range(nVertices)]
    wt[0] = 0
    visited = [False for _ in range(nVertices)]
    unvisited = [i for i in range(nVertices)]
    
    res = []
    for _ in range(nVertices-1):
        v = getMinWeightedVertex(wt, unvisited)
        visited[v] = True
        unvisited.remove(v)
        
        for j in range(nVertices):
            if adjMatrix[v][j] and not visited[j]:
                if adjMatrix[v][j] < wt[j]:
                    wt[j] = adjMatrix[v][j]
                    parent[j] = v
                    
    for i in range(1, nVertices):
        if i < parent[i]:
            res.append([i, parent[i], wt[i]])
        else:
            res.append([parent[i], i, wt[i]])
    
    return res
        
        
        
        
        
        






nV, nE = input().split()


adjMatrix = [[0 for _ in range(int(nV))] for _ in range(int(nV))]
for _ in range(int(nE)):
    v1, v2, wt = input().split()
    adjMatrix[int(v1)][int(v2)] = int(wt)
    adjMatrix[int(v2)][int(v1)] = int(wt)
    
res = prims(adjMatrix, int(nV))
for i in res:
    print(*i)
        
'''input'''       
# 5 7
# 0 1 4
# 0 2 8
# 1 2 2
# 1 3 6
# 2 3 3
# 2 4 9
# 3 4 5

'''output'''
# 0 1 4
# 1 2 2
# 2 3 3
# 3 4 5
```

# Dijkstra's Algorithm

```python
def getMinDisVertex(dis, unvisited):
    v = unvisited[0]
    minimum = dis[v]
    for i in unvisited:
        if dis[i] < minimum:
            minimum = dis[i]
            v = i
            
    return v


def dijkstras(adjMatrix, nVertices):
    dis = [100000 for _ in range(nVertices)]
    dis[0] = 0
    visited = [False for _ in range(nVertices)]
    unvisited = [i for i in range(nVertices)]
    
    res = []
    for _ in range(nVertices):
        v = getMinDisVertex(dis, unvisited)
        visited[v] = True
        unvisited.remove(v)
        
        for j in range(nVertices-1):
            if adjMatrix[v][j] and not visited[j]:
                if dis[v] + adjMatrix[v][j] < dis[j]:
                    dis[j] = dis[v] + adjMatrix[v][j]
                
    for i in range(nVertices):
        res.append([i, dis[i]])
        
        
    return res
        
        
        
        
        
nV, nE = input().split()


adjMatrix = [[0 for _ in range(int(nV))] for _ in range(int(nV))]
for _ in range(int(nE)):
    v1, v2, wt = input().split()
    adjMatrix[int(v1)][int(v2)] = int(wt)
    adjMatrix[int(v2)][int(v1)] = int(wt)
    
res = dijkstras(adjMatrix, int(nV))
for i in res:
    print(*i)
```
